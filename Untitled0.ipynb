```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.axes as ax


# Total de acidentes com vítima por bairro em acidentes com embriaguez;



dF1 = pd.read_csv("drive/MyDrive/TrabalhoFinalBigData/si_envolvidos_2019.csv", sep=';', header=0, index_col = None ,encoding='latin_1')
dF2 = pd.read_csv("drive/MyDrive/TrabalhoFinalBigData/si_logradouro_2019.csv", sep=';', header=0, index_col = None  ,encoding='latin_1')
dF3 = pd.read_csv("drive/MyDrive/TrabalhoFinalBigData/si_boletimOcorrencia_2019.csv", sep=';', header=0, index_col = None  ,encoding='latin_1')

dF4 = pd.merge(dF1, dF2, left_on="num_boletim", right_on="Nº_boletim")

ax = dF5 = dF4.loc[(dF4[' Embreagues'] == 'SIM')]

ax = dF_pie = dF5[['nome_bairro', ' Nº_envolvido']].groupby('nome_bairro', as_index=True).sum()

ax = dF_pie.plot.barh(y=' Nº_envolvido', figsize=(15, 30), color='red')

ax.set_title('Total de acidentes com vítima por bairro em acidentes com embriaguez',fontsize = 20,color='red')

ax.legend(bbox_to_anchor=(1,1), loc=0, borderaxespad=0.8, fontsize = 15)



```

```python
import pandas as pd
import matplotlib.axes as ax
# Total de acidentes por tipo de pavimento e condição do tempo;


dF1 = pd.read_csv("drive/MyDrive/TrabalhoFinalBigData/si_envolvidos_2019.csv", sep=';', header=0, index_col = None ,encoding='latin_1')
dF2 = pd.read_csv("drive/MyDrive/TrabalhoFinalBigData/si_logradouro_2019.csv", sep=';', header=0, index_col = None  ,encoding='latin_1')
dF3 = pd.read_csv("drive/MyDrive/TrabalhoFinalBigData/si_boletimOcorrencia_2019.csv", sep=';', header=0, index_col = None  ,encoding='latin_1')



dF = pd.merge(dF1, dF3, left_on="num_boletim", right_on=" NUMERO_BOLETIM")

ax = dF_pie = dF[[' PAVIMENTO',' DESC_TEMPO',' Nº_envolvido']].groupby([' PAVIMENTO',' DESC_TEMPO'], as_index=True).sum()

ax = dF_pie.plot.barh(y=' Nº_envolvido', figsize=(15, 15),colormap='inferno')

ax.set_title('Total de acidentes por tipo de pavimento e condição do tempo',fontsize = 20,color='b')

ax.legend(bbox_to_anchor=(0.9,1), loc=0, borderaxespad=0.8, fontsize = 15)


```

```python
import pandas as pd
import matplotlib.axes as ax
# Total de pessoas acidentadas por tipo de veiculo e tipo de pavimentação;

dF1 = pd.read_csv("drive/MyDrive/TrabalhoFinalBigData/si_envolvidos_2019.csv", sep=';', header=0, index_col = None ,encoding='latin_1')
dF2 = pd.read_csv("drive/MyDrive/TrabalhoFinalBigData/si_logradouro_2019.csv", sep=';', header=0, index_col = None  ,encoding='latin_1')
dF3 = pd.read_csv("drive/MyDrive/TrabalhoFinalBigData/si_boletimOcorrencia_2019.csv", sep=';', header=0, index_col = None  ,encoding='latin_1')

dF = pd.merge(dF1, dF3, left_on="num_boletim", right_on=" NUMERO_BOLETIM")

ax = dF1 = dF.loc[(dF['especie_veiculo'] != '') & (dF[' PAVIMENTO'] != '')]

ax = dF_pie = dF[['especie_veiculo',' PAVIMENTO',' Nº_envolvido']].groupby(['especie_veiculo',' PAVIMENTO'], as_index=True).sum()

ax = dF_pie.plot.barh(y=' Nº_envolvido', figsize=(15, 15),colormap='summer')

ax.set_title('Total de pessoas acidentadas por tipo de veiculo e tipo de pavimentação',fontsize = 20,color='green')

ax.legend(bbox_to_anchor=(0.9,1), loc=0, borderaxespad=0.8,fontsize = 15)

```

```python
import pandas as pd
# Média de idade dos condutores por tipo de veículo e tipo de acidente;

dF1 = pd.read_csv("drive/MyDrive/TrabalhoFinalBigData/si_envolvidos_2019.csv", sep=';', header=0, index_col = None ,encoding='latin_1')
dF2 = pd.read_csv("drive/MyDrive/TrabalhoFinalBigData/si_logradouro_2019.csv", sep=';', header=0, index_col = None  ,encoding='latin_1')
dF3 = pd.read_csv("drive/MyDrive/TrabalhoFinalBigData/si_boletimOcorrencia_2019.csv", sep=';', header=0, index_col = None  ,encoding='latin_1')

dF = pd.merge(dF1, dF3, left_on="num_boletim", right_on=" NUMERO_BOLETIM")

ax = dFvar = dF.loc[(dF[' condutor'] == 'S') ]

ax = df4 = dFvar[[' DESC_TIPO_ACIDENTE',' Idade']].groupby(' DESC_TIPO_ACIDENTE', as_index=True).mean()

ax = df4.plot.barh(y=' Idade', figsize=(15, 15), fontsize = 15, colormap = 'coolwarm')

ax.set_title('Média de idade dos condutores por tipo de veículo e tipo de acidente',fontsize = 20,color='blue')

ax.legend(bbox_to_anchor=(1,1), loc=0, borderaxespad=0.8,fontsize = 15)
```

```python
import pandas as pd
# Média de idade dos condutores por indicativo de embriaguez;

dF1 = pd.read_csv("drive/MyDrive/TrabalhoFinalBigData/si_envolvidos_2019.csv", sep=';', header=0, index_col = None ,encoding='latin_1')
dF2 = pd.read_csv("drive/MyDrive/TrabalhoFinalBigData/si_logradouro_2019.csv", sep=';', header=0, index_col = None  ,encoding='latin_1')
dF3 = pd.read_csv("drive/MyDrive/TrabalhoFinalBigData/si_boletimOcorrencia_2019.csv", sep=';', header=0, index_col = None  ,encoding='latin_1')

dFvar = dF1.loc[(dF1[' condutor'] == 'S') ]

ax = df4 = dFvar[[' Embreagues',' Idade']].groupby(' Embreagues', as_index=True).mean()

ax = df4.plot.barh(y=' Idade', figsize=(15, 15), colormap = 'autumn')

ax.set_title('Média de idade dos condutores por indicativo de embriaguez',fontsize = 20,color='red')

ax.legend(bbox_to_anchor=(1,1), loc=0, borderaxespad=0.8,fontsize = 15)

```
